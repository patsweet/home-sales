```python
# General Imports
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress
# Enable inline plotting
%matplotlib inline
```

### Preparing the data

```python
# Grab New Castle property and add a 'county' column.
ncco = pd.read_csv('data/New Castle Home Sales.csv')
ncco['county'] = 'N'

# Grab Kent property data and add a county column.
kent = pd.read_csv('data/Kent Home Sales.csv')
kent['county'] = 'K'

# Combine the datasets and filter to only years with 5+ sales.
df = ncco.append(kent)
print "Number of subdivisions: %d" % len(df.Subdiv.unique())
df = df[df.Sales >= 5]
print "Number of subdivisions after filtering: %d" % len(df.Subdiv.unique())
df = df.set_index(['Subdiv'])

# Group by Subdivision for later mucking about
by_subdiv = df.groupby(df.index)
```

## Examining Median Sold Price vs. Median Days on Market

```python
def scatterPlusTrend(xval, yval):
    """Create a scatterplot with a trendline using
    scipy's linregress function to perform the least
    squares regression."""
    plt.figure(figsize=(15,8))
    plt.grid()
    slope, intercept, r_value, p_value, std_err = linregress(x=df[xval], y=df[yval])
    print 'R-Squared: {:.5f}'.format(r_value**2)
    print 'P Value: {:.5f}'.format(p_value)
    print 'Standard Error: {:,.2f}'.format(std_err)
    plt.scatter(x=df[xval], y=df[yval])
    plt.plot(df[xval], slope * df[xval] + intercept, 'r-')
```

```python
scatterPlusTrend('Median DOM', 'Median Sold Price')
```

```python
def histogramPlusAgg(var):
    """Produces a histrogram based on the provided variable
    along with very basic aggregate stats - mean, max and min -
    formatted to 4 decimal places."""
    print df[var].hist(bins=60,figsize=[15,8])
    print "Average {}: {:,.4f}".format(var, df[var].mean())
    print "Max {}: {:,.4f} ({!s})".format(
        var,
        df[var].max(),
        df[df[var] == df[var].max()].index[0]
    )
    print "Min {}: {:,.4f} ({})".format(
        var,
        df[var].min(),
        df[df[var] == df[var].min()].index[0]
    )
    
histogramPlusAgg('Median Sold Price')
```

## How does Days on Market impact Sale Price/Original Listing price?

```python
scatterPlusTrend('Median DOM', 'Median.SP.OLP')
```

There doesn't seem to be too strong of a correlation between Days on Market and the Sale Price/Original List Price ratio. The least squares regression shows that the longer the home is on the market, the least likely the seller is to get the original listing price. However, the r-squared produced by the regression suggests the relationship is not very strong.

```python
histogramPlusAgg('Median.SP.OLP')
```
