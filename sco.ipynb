# <a name='top'></a>Analysis of Sussex County property data by subdivision

In this excercise, I take subdivision level property data and try to answer the following questions:

1. [In which subdivisions are houses selling the fastest? Selling slowest?](#q1)
2. [In which subdivisions is the median sale price improving most? Dropping most?](#q2)
3. [Where can a seller get the best deal in terms of median sale price / median list price](#q3)

## Methodology

To tackle these questions, I'm using [Pandas](http://pandas.pydata.org/) - a Python library for data analysis - and [SciPy](http://docs.scipy.org/doc/scipy-0.13.0/reference/index.html) - a mathematics, science and engineering library - to filter, group and ultimately run a [ordinary least squares linear regression](http://en.wikipedia.org/wiki/Ordinary_least_squares) to determine the positive or negative movement on a number of variables.

## About the data

The town/ZIP code level data comes from the InnoVia MLS and provided by the Sussex County Association of Realtors. Unfortunately, the provided data does not include all of the fields that were used in similar anayses for the other two counties. Specifically, we are missing `Median DOM`, `Median Sale Price/Original Listing Price` and `Median Sale Price/Final Listing Price`. We do have a Median Sale Price/Median Final Listing Price for each area - a much more genralized figure than the other measures: Here is the basic structure of our data:

 - Subdivision
     - The name of the subdivision
 - Year
     - The data year
 - Sales
     - Number of transactions
     - (Note this doesn't necessarily mean number of homes)
 - Median Sale Price
    - Median price of transactions
 - Avg. Sale Price
     - Average price of transactions
 - Avg. Days on Market
     - Average number of days the house was listed for sale.
 - Median Sale Price / Median Listing Price
     - The ratio between the median sale price for the total area and the median listing price for the total area.
     - A value above 1.0 means the median sale price for the area was above the median listing price.

## Known Issues

There are three issues that directly affect this analysis:

1. We do not have data for the entire county. The Sussex County Association of Realtors was only able to provide us with data for the beach communities and select communities in the western part of the county; 25 communities in total. The reasoning, we were told, is that there system is quite limited and they were only able to pull out data for one community at a time (ouch).
2. Averages. Unlike data for the other counties, we only have `Average DOM`.
3. The figures we have for Sale Price vs. Listing Price are at the area level, not the transaction level. It's simply the median sale price for the entire area divided by the median listing price for the entire area.

## Preparing the data

```python
# General Imports
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress
import re
# Enable inline plotting
%matplotlib inline
```

```python
# import the data into a Pandas dataframe.
df = pd.read_csv('data/Sussex Home Sales.csv')
print "Original Data:"
print df.info()
print "="*100

# Filter out data where the subdivision is "None Available."
df = df[df.Subdiv != 'None Available']

# Get the total number of subdivisions
orig_subdiv_count = len(df.Subdiv.unique())

# Filter out data where there are less than 5 sales
df = df[df.Sales >= 5]

# Filter out the subdivisions that don't have 4 years
# worth of data.
df = df.groupby('Subdiv').filter(lambda x: x.count() >= 3)

# Get the number of subdivisions that survived the cuts
new_subdiv_count = len(df.Subdiv.unique())

# R-SQUARED THRESHOLD
R2_THRESHOLD = 0.5

# Set the Subdivision as the Index.
df = df.set_index(['Subdiv'])
print "Filtered Data:"
df.info()

# Group by the subdivisions for use in our calculations below
by_subdivision = df.groupby(df.index)
```

# 1. <a name='q1'></a>In which subdivisions are houses selling the fastest? Selling slowest?

Grouping our data by subdivision, we are able to run each through a linear regression with SciPy's `linregress` function.

[Back to top](#Analysis-of-New-Castle-County-property-data-by-subdivision)

```python
def calculateSlope(df, attribute, r_squared_threshold=R2_THRESHOLD):
    """Gets the slope of an attribute of the dataframe along with
    min, max and r-squared values."""
    # Using scipy's linregress, get the slope of dom/year for each subdivision
    slope = pd.DataFrame(df.apply(lambda x: linregress(y=x[attribute], x=x.Year)[0]))
    r2 = pd.DataFrame(df.apply(lambda x: linregress(y=x[attribute], x=x.Year)[2]**2))
    output = pd.merge(left=slope, right=r2, left_index=True, right_index=True)
    
    # Get the max and min DOM for each subdivision.
    output = pd.merge(left=output, right=pd.DataFrame(df[attribute].min()), left_index=True, right_index=True)
    output = pd.merge(left=output, right=pd.DataFrame(df[attribute].max()), left_index=True, right_index=True)
    output= pd.merge(left=output, right=pd.DataFrame(df['Sales'].sum()), left_index=True, right_index=True)
    
    # Fix Column names
    attribute = re.sub(r'[\W]+', '_', attribute).lower()
    output.columns = [attribute+'_slope', attribute+'_r2', attribute+'_min', attribute+'_max', 'sales_count']
    
    return output, output[output[attribute+'_r2'] >= r_squared_threshold]
```

```python
average_dom, average_dom_adj = calculateSlope(by_subdivision, 'Average DOM')
average_dom[['average_dom_slope', 'average_dom_r2']].hist(bins=15, figsize=[14,6])
```

```python
print "The best fit:"
average_dom[average_dom.average_dom_r2 == average_dom.average_dom_r2.max()]
```

```python
print "The Worst fit:"
average_dom[average_dom.average_dom_r2 == average_dom.average_dom_r2.min()]
```

### Throwing it all together for output to csv

```python
# Combine all of our stats into a single dataframe.
merge_out = pd.merge(
    left=average_dom[['sales_count','average_dom_slope', 'average_dom_r2', 'average_dom_min', 'average_dom_max']],
    right=median_saleprice[['median_sold_price_slope', 'median_sold_price_r2', 'median_sold_price_min', 'median_sold_price_max']],
    left_index=True,
    right_index=True)
merge_out = pd.merge(
    left=merge_out,
    right=lp[['median_sp_lp_slope', 'median_sp_lp_r2', 'median_sp_lp_min', 'median_sp_lp_max']],
    left_index=True,
    right_index=True)

# Add the rankings for each.
merge_out['dom_slope_rank'] = merge_out['average_dom_slope'].rank(ascending=True)
merge_out['median_price_rank'] = merge_out['median_sold_price_slope'].rank(ascending=False)
merge_out['sp_lp_rank'] = merge_out['median_sp_lp_slope'].rank(ascending=False)
merge_out['average_rank'] = merge_out[['dom_slope_rank', 'median_price_rank', 'sp_lp_rank']].mean(axis=1)
print "The Top Ten"
print merge_out.sort('average_rank')[['average_rank']].head(10)


merge_out.sort('average_rank').to_csv('data_out_sco.txt', sep='\t')
```
